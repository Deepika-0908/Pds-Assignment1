```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load cleaned data
cleaned_data = pd.read_csv('/content/data_clean/cleaned_student_performance.csv')

# Create a folder for visualizations
output_folder = '/content/visualizations'
os.makedirs(output_folder, exist_ok=True)

# Visualization 1: Histogram of Math Scores
plt.figure(figsize=(8, 6))
sns.histplot(cleaned_data['math score'], kde=True, color='blue')
plt.title("Distribution of Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Frequency")
plt.savefig(os.path.join(output_folder, 'visualization_1.png'))
plt.show()

# Visualization 2: Scatter Plot of Math vs. Reading Scores
plt.figure(figsize=(8, 6))
sns.scatterplot(x='math score', y='reading score', data=cleaned_data, hue='gender')
plt.title("Math Scores vs. Reading Scores")
plt.xlabel("Math Score")
plt.ylabel("Reading Score")
plt.savefig(os.path.join(output_folder, 'visualization_2.png'))
plt.show()

# Visualization 3: Box Plot of Writing Scores by Parental Education Level
plt.figure(figsize=(10, 6))
sns.boxplot(x='parental level of education', y='writing score', data=cleaned_data, hue='parental level of education', palette='Set2', legend=False)
plt.title("Writing Scores by Parental Education Level")
plt.xlabel("Parental Education Level")
plt.ylabel("Writing Score")
plt.xticks(rotation=45)
plt.savefig(os.path.join(output_folder, 'visualization_3.png'))
plt.show()

# Visualization 4: Bar Plot of Average Scores by Gender
avg_scores = cleaned_data.groupby('gender')[['math score', 'reading score', 'writing score']].mean()
avg_scores.plot(kind='bar', figsize=(10, 6))
plt.title("Average Scores by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.savefig(os.path.join(output_folder, 'visualization_4.png'))
plt.show()

# Visualization 5: Heatmap of Correlation Matrix
plt.figure(figsize=(8, 6))
correlation_matrix = cleaned_data[['math score', 'reading score', 'writing score']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix of Scores")
plt.savefig(os.path.join(output_folder, 'visualization_5.png'))
plt.show()
```
